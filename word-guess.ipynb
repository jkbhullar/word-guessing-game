```python
    
def getWord():
    global dictionary
    global roundWord
    
    
    roundUnderscoreWord = []
    blank = "_"

    roundWord = random.choice(dictionary) 
    

    for i in range(0, len(roundWord)):
        roundUnderscoreWord.append(blank)
    
    
    return roundWord, roundUnderscoreWord 
getWord()
```

```python
def wofRoundSetup():
    global players
    global roundWord
    global blankWord
    
   
    
    for playerNum in range(0, len(players)):
        players[playerNum]["roundtotal"] = 0
    
    initPlayer = random.choice(list(players))
    playerNum = initPlayer
 
    roundWord, blankWord = getWord()
#     debugging purposes:
#     print(list(roundWord))
    print(f'{players[playerNum]["name"]}, it is your turn. Here is the board: {blankWord}')
    

    return initPlayer
wofRoundSetup()
```

```python
def spinWheel(playerNum):
    global wheellist
    global players
    global wheelSpin
    global roundWord
    

    wheel = [100, 150, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650, 700, 750, 800, 850, 900]
    wheelSpin = random.choice(wheel)
    
    print(f'You rolled: {wheelSpin}')
    
    
    letter = input("Guess a letter: ")
    goodGuess = guessletter(letter)
    if goodGuess:
        players[playerNum]["roundtotal"] += int(wheelSpin)
        print(f'{players[playerNum]["name"]}, your round total is now: {players[playerNum]["roundtotal"]}')
    return goodGuess
    

spinWheel(0)
```
